# Assignment 1: Data Cleaning (Gender Column)
This notebook loads the reduced Emergency Triage dataset. It deals with missing values in the "Gender" column by dropping their corresponding rows. Additionally, it maps the different unique values of the "Gender" column to two values: `1` (Male) and `0` (Female).

```python
# Confirm Python version >= 3.10
import sys
print(f'Python version: {sys.version}')
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np

print('Libraries imported successfully!')
```

```python
# Load dataset and display shape & column names
FILE_PATH = 'EmergencyTriageDataset_Reduced_Dirty.csv'
df_raw = pd.read_csv(FILE_PATH)

print('Dataset successfully loaded!')
print(f'Shape: {df_raw.shape[0]} rows x {df_raw.shape[1]} columns')
print(f'Columns: {list(df_raw.columns)}')
```

```python
# Display first 10 rows
df_raw.head(10)
```

```python
# Display the data types of each column
print(df_raw.dtypes)
```

```python
# Check for any missing values in the 'Gender' column
print(f'Number of NaN values in "Gender": {df_raw['Gender'].isnull().sum()}')
```

**Note:** No cases of missing values in the "Gender" column so no action is required to rectify this

```python
# Drop rows with NaN value
df_raw = df_raw.dropna(subset=['Gender'])
```

```python
# Display all the unique values in the 'Gender' column
print(f'Unique values in "Gender": {df_raw['Gender'].unique()}')
print(f'Number of unique values in "Gender": {df_raw['Gender'].nunique()}')
```

```python
# Display 'Gender' value counts
print(f'Gender value counts: {df_raw['Gender'].value_counts()}')
```

```python
# Build value map
gender_map = {
    'male': 1,
    '1': 1,
    'female': 0,
    '0': 0
}
```

```python
# Normalize all values to lower case
df_raw['Gender_Cleaned'] = df_raw['Gender'].astype(str).str.lower().str.strip().map(gender_map)
```

```python
# Display value counts for new columns
print(f'Cleaned "Gender" value counts: {df_raw['Gender_Cleaned'].value_counts()}')
```

```python
# Check for null values in the cleaned column
print(f'Number of NaN values in "Gender": {df_raw['Gender_Cleaned'].isnull().sum()}')
```

```python
# Drop old "Gender" column and rename cleaned column
df_raw = df_raw.drop(columns=['Gender'])
df_raw = df_raw.rename(columns={'Gender_Cleaned': 'Gender'})

df_raw.head(10)
```
